# Pitcher IP Distribution — Phase 4 Validation (Corrected)

Validates whether calibrated PT-model IP with statcast-gbm direct rate stats produces
better pitcher valuations than the steamer-IP baseline.

**Fix applied:** Phase 4 originally used `--param playing_time=playing_time` which does NOT
route IP through the playing_time model in routed mode. The consensus PT override is only
applied in `weighted_average` mode. The corrected approach adds `playing_time` as a normal
ensemble component and routes `ip`/`pa` to it via per-stat route overrides:

```
--param "components.playing_time=1.0"
--param "versions.playing_time=latest"
--param "routes.ip=playing_time"
--param "routes.pa=playing_time"
```

This was verified: ensemble IP now matches playing_time predictions exactly (not steamer).

**Systems compared:**
- `steamer-baseline`: Steamer-only projections (ZAR/SGP)
- `direct-rate-steamer`: Routed ensemble with steamer IP + statcast-gbm direct rates (current best)
- `pt-direct-rate`: Routed ensemble with PT-model IP + statcast-gbm direct rates (candidate) — **correctly routed**

**Holdout seasons:** 2024, 2025

### Corrected results vs original (broken routing)

The corrected routing dramatically changes the picture vs `direct-rate-steamer`:

- **Original (broken):** Failed 3 of 4 comparisons vs direct-rate-steamer (only ZAR 2024 passed)
- **Corrected:** Passes all 4 comparisons vs direct-rate-steamer — WAR ρ improves by +0.07 to +0.29 across the board

The improvement is most dramatic for pitcher WAR ρ, which makes sense: the broken routing was using steamer IP (identical to the baseline), so there was no actual PT-model signal in the comparison.

### Key observations

1. **Dominates direct-rate-steamer on both seasons** — WAR ρ (all) improves by +0.29 (2024) and +0.11 (2025) for ZAR; +0.32 (2024) and +0.17 (2025) for SGP. Hit rates improve substantially. This is a clear upgrade over steamer IP + direct rates.

2. **Mixed vs steamer-baseline** — Passes on 2024 but fails on 2025. The 2025 failure is driven by pitcher WAR ρ dropping by 0.09-0.11, though value MAE and rank ρ improve.

3. **Top-10 concentration is too high** — Both seasons show top pitcher > 2x the 10th pitcher ($141.6/$39.6=3.57x in 2024, $100.6/$38.8=2.60x in 2025). Paul Skenes at $141.6 in 2024 is especially extreme.

4. **Value calibration consistently improves** — Value MAE drops by $0.27-$1.70 across all comparisons and seasons. Rank ρ also consistently improves.

5. **Pitcher WAR ρ is the weak point** — While overall WAR ρ improves (driven by batters), pitcher-specific WAR ρ regresses vs steamer-baseline on 2025 and drops vs steamer-baseline SGP on 2024.

## Decision: NO-GO (confirmed after routing fix)

The PT-model IP distribution with direct rate stats shows **significant improvement over direct-rate-steamer** (the current best with steamer IP), but still does **not** fully pass the validation criteria:

- **PASS** all 4 comparisons vs direct-rate-steamer (both seasons, both systems)
- **FAIL** vs steamer-baseline on 2025 — pitcher WAR ρ drops 0.09-0.11
- **FAIL** top-10 pitcher concentration exceeds 2x threshold on both seasons (3.57x and 2.60x)
- **PASS** value MAE and rank ρ consistently improve across all comparisons

### What changed with the routing fix

The original phase 4 validation used `--param playing_time=playing_time`, which only works in
`weighted_average` mode. In `routed` mode, IP was silently coming from steamer — making the
"PT-model IP" candidate identical to the steamer-IP baseline for pitcher IP. The corrected
routing (`routes.ip=playing_time`) actually delivers PT-model IP, and the results improve
dramatically vs the direct-rate-steamer baseline.

### Why still NO-GO

1. **Top-10 concentration (3.57x, 2.60x)** far exceeds the 2x threshold. The PT-model IP
   distribution inflates elite starter values excessively.
2. **Pitcher WAR ρ regresses on 2025** vs steamer-baseline, suggesting the PT model's IP
   predictions for 2025 pitchers are less accurate than steamer's.

**Production config remains unchanged** — continue using steamer IP with direct rates.

### Future work

The corrected results are much more promising than the original broken validation suggested.
The path forward is:
1. **Fix top-10 concentration** — cap or dampen extreme IP predictions at the top
2. **Investigate 2025 pitcher WAR ρ regression** — identify which pitchers the PT model
   gets wrong in 2025 that steamer gets right
3. **Consider a hybrid approach** — use PT-model IP for the middle of the distribution
   (where it excels at value calibration) but blend with steamer at the extremes
