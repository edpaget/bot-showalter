## Summary

| Comparison | Season | ZAR | SGP |
|-----------|--------|-----|-----|
| vs steamer-baseline | 2024 | **PASS** | **PASS** |
| vs steamer-baseline | 2025 | FAIL | FAIL |
| vs direct-rate-steamer | 2024 | **PASS** | FAIL |
| vs direct-rate-steamer | 2025 | FAIL | FAIL |

### Key observations

1. **2024 shows improvement over steamer-only** — WAR ρ jumps from 0.20 to 0.30 (ZAR), and pitcher WAR ρ improves substantially (+0.11 ZAR, though SGP pitcher ρ drops).

2. **2025 consistently fails** — Pitcher WAR ρ drops by 0.10-0.14 across both valuation systems vs both baselines. The PT-model IP distribution does not generalize well to 2025.

3. **vs direct-rate-steamer (current best)** — Even in the favorable 2024 season, the PT candidate fails SGP and only passes ZAR. In 2025 it fails both systems with large regressions in hit rate top-100 (-14pp ZAR, -19pp SGP).

4. **Top-10 concentration is too high** — Both seasons show top pitcher > 2x the 10th pitcher ($117/$45.9=2.55x in 2024, $103/$47.6=2.16x in 2025), indicating the IP distribution inflates elite starter values excessively.

5. **Value MAE improves** — PT-direct-rate consistently has lower value MAE (better calibrated dollar values), but this doesn't translate to better ranking accuracy.

### Root cause hypothesis

The PT model narrows the pitcher pool (from ~6000 to ~950 valuated players), which improves value calibration (lower MAE) but distorts replacement level. The narrower pool means the replacement baseline is set by better pitchers, making it harder to identify value above replacement correctly. This is why hit rates drop especially at top-100 — marginal pitchers around the roster cutoff get incorrectly valued.

## Decision: NO-GO

The PT-model IP distribution with direct rate stats does **not** pass the validation criteria:

- ❌ Pitcher WAR ρ regresses by >0.01 on 2025 (drops 0.05-0.14 depending on system/baseline)
- ❌ Top-10 pitcher concentration exceeds 2x threshold on both seasons
- ❌ Fails `--check` on 2025 against both steamer-baseline and direct-rate-steamer
- ✅ Passes `--check` on 2024 vs steamer-baseline only
- ✅ Value MAE consistently improves

**Production config remains unchanged** — continue using steamer IP with direct rates (the `direct-rate-steamer` configuration) as the production valuation setup.

### Future work

The PT-model IP approach could be revisited if:
1. The isotonic calibration is retrained with a broader calibration window (currently uses limited seasons)
2. The replacement level calculation is adjusted to account for the narrower pitcher pool
3. A hybrid approach blends PT-model IP for starters with steamer IP for relievers
